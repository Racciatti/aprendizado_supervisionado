**(b) Para as variáveis selecionadas em (a), utilize funções de posição relativa para:**

* Obter o intervalo interquartílico (IQR).

```python
selected_variables = 'AgeStartedSmokingCigarettesRegularly DaysSmokedCigsDuringPast30Days CigarettesSmokedPerDay DaysSmokedCigarettesLast5Days'.split()
df1 = selected_samples_df[selected_variables]
```

```python
df1.info()
```

```python
stats = {}

for var_name in selected_variables:
    stats[var_name] = {}
    stats[var_name]['mean'] = df1[var_name].mean()
    stats[var_name]['median'] = df1[var_name].median()
    stats[var_name]['std'] = df1[var_name].std()
    stats[var_name]['max'] = df1[var_name].max()
    stats[var_name]['min'] = df1[var_name].min()
    stats[var_name]['amplitude'] = df1[var_name].min()
    stats[var_name]['z-score'] = stats[var_name]['std'] / stats[var_name]['mean']
    stats[var_name]['Q1'] = df1[var_name].quantile(0.25)
    stats[var_name]['Q2'] = df1[var_name].quantile(0.5)
    stats[var_name]['Q3'] = df1[var_name].quantile(0.75)
    stats[var_name]['IQR'] = stats[var_name]['Q3'] - stats[var_name]['Q1']
    
    
    
    
```

```python
for var_name in stats:
    print(f"=== {var_name} ===")
    for stat_name in stats[var_name]:
        print(f"  {stat_name}: {stats[var_name][stat_name]}")
    print('\n')
```

**Explicação de como medidas baseadas em quantis contribuem**: Medidas em quantis são essencialmente uma estatística que nos informa a distribuição dos dados de forma mais rica que métricas tradicionais, como a média e o desvio padrão. 

Os quantis funcionam com base na seguinte lógica:
1. Os valores em um determinado vetor são ordenados
2. A frequência relativa de cada valor presente é calculada
3. O vetor é dividido em dois subconjuntos, tomando como base um valor $Q\in[0,1]$, onde $Q$ representa um determinado ponto da distribuição dos dados, do menor para o maior valor, em relação ao número total de amostras.
    * Por exemplo: Se $Q=0.5$, o valor resultante será aquele que divide o conjunto original em dois subconjuntos com 50% do número total de amostras (neste caso, a mediana).
    * Se $Q=0.2$, o valor resultante será o valor da distribuição que é maior que 20% e menor que 80% dos valores restantes. 

**Importância**: Com base em um número relativamente de valores para $Q$ (e.g. $[0.2, 0.4, 0.6, 0.8]$), podemos ter uma visão mais detalhada da distribuição dos dados (onde se concentram, onde são esparsos, etc.).

**Exemplificando**: Se 
* para $Q = 0$ temos o valor 0 (menor valor do vetor)
* para $Q = 1$ temos o valor 1000 (menor valor do vetor)
* para $Q = 0.25$ temos o valor 0.1 (menor valor do vetor)
* para $Q = 0.5$ temos o valor 1.5 (menor valor do vetor)
* para $Q = 0.75$ temos o valor 3.1 (menor valor do vetor)

Podemos concluir que a distribuição possui uma cauda extremamente longa à direita e possivelmente outliers (alta densidade de amostras para valores pequenos da variável sendo observada).

##### Discussão sobre a escolha do particionamento (bins)
A escolha do número de intervalos para contagem de frequência no histograma é um trade-off entre detalhamento (maior granularidade, intervalos menores) e simplificação / aproximação.

Neste caso, a maior granularidade parecer ser melhor (sturges, 50 bins), devido as variáveis apresentarem uma distribuição relativamente comum (parece uma gaussiana ou binomial).

No entanto para dados com uma distribuição mais caótica, como por exemplo o horário de voos em um dado aeroporto, seria interessante optar por uma granularidade menor para a observação de padrões (e.g. ao invés de uma coluna para cada hora de voos, colocar uma coluna por semana ou mesmo por mês para observar sazonalidade)

```python
# Married data
married_df = box_df[box_df['MaritalStatus'] == 'Married']
# Never married data
never_married_df = box_df[box_df['MaritalStatus'] == 'Never married']
# Separated data
separated_df = box_df[(box_df['MaritalStatus'] == 'Separated') | (box_df['MaritalStatus'] == 'Divorced') | (box_df['MaritalStatus'] == 'Widowed')]

print(f"{married_df.info()}")
print(f"{never_married_df.info()}")
print(f"{separated_df.info()}")
```

```python
data = [
    married_df['TotalPercentFat'],
    never_married_df['TotalPercentFat'],
    separated_df['TotalPercentFat']
]

labels = [
    'Married',
    'Never Married',
    'Separated'
]

# 1. Define styling dictionaries for high customization  (ALL STYLING WAS AI-GENERATED)
box_style = dict(facecolor='#e6f2ff', color='#0052cc', linewidth=2)
whisker_style = dict(color='#0052cc', linewidth=1.5, linestyle='--')
median_style = dict(color='#ff3300', linewidth=2)
cap_style = dict(color='#0052cc', linewidth=1.5)
fliers_style = dict(marker='o', markerfacecolor='#ff3300', markersize=5)

fig, ax = plt.subplots(figsize=(10, 6))

# 2. Apply styles to the boxplot
ax.boxplot(
    data,
    patch_artist=True,    # Required to enable box filling (facecolor)
    tick_labels=labels,   # Set custom labels for each box
    notch=True,           # Adds a notch for the median confidence interval
    boxprops=box_style,
    whiskerprops=whisker_style,
    medianprops=median_style,
    capprops=cap_style,
    flierprops=fliers_style
)

# 3. Final visual polish
ax.yaxis.grid(True, linestyle='--', alpha=0.6)  # Add horizontal grid lines
ax.set_title('Total Percent Body Fat by Marital Status', fontsize=14, pad=15)
ax.set_ylabel('Total Percent Fat (%)', fontsize=12)

plt.tight_layout()
plt.show()
```


##### **(c) Gráfico de Violino com ggplot2.**

* Reproduza a comparação do item (b) usando ggplot() e geom_violin(), inclua rótulos informativos com labs() e ajuste um tema com theme().
* Discuta, em termos exploratórios, quais informações adicionais o gráfico de violino fornece em relação ao boxplot.


```python
import matplotlib.pyplot as plt
data = [
    married_df['TotalPercentFat'], 
    never_married_df['TotalPercentFat'], 
    separated_df['TotalPercentFat']
]

labels = ['Married', 'Never Married', 'Separated']

fig, ax = plt.subplots(figsize=(10, 6))


parts = ax.violinplot(data, showmeans=False, showmedians=True)

# 2. Stylize the "bodies" (the filled areas) (AI - GENERATED)
for pc in parts['bodies']:
    pc.set_facecolor('#0052cc')  # Fill color
    pc.set_edgecolor('black')    # Outline color
    pc.set_alpha(0.7)            # Transparency

# 3. Stylize the internal lines (median, max, min, etc.)
parts['cmedians'].set_color('#ff3300') # Median line color
parts['cmedians'].set_linewidth(2)
parts['cmins'].set_color('#0052cc')
parts['cmaxes'].set_color('#0052cc')
parts['cbars'].set_color('#0052cc')    # The vertical center bar

# 4. Final visual polish
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(labels)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
ax.set_title('Distribution of Total Percent Fat by Marital Status', fontsize=14)
ax.set_ylabel('Total Percent Fat (%)', fontsize=12)

plt.tight_layout()
plt.show()
```
